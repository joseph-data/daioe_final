# Employment Weighted DAIOE

This notebook documents the employment-weighted DAIOE aggregation pipeline and calls the reusable helpers from `scripts/02_weighting.py`.

- **Data lineage**: DAIOE raw subindices (tab-separated) are merged with the latest SCB employment snapshot for the same taxonomy.
- **Hierarchy terms**: `code4` values are the finest SSYK occupations; levels 1–3 are progressively broader groups. `n_children` counts how many descendants a code has in the next level down.

## Formulas

Employment-weighted aggregation for a group `g` (levels 1–3):

$$D^{(w)}_g = \frac{\sum_{k \in g} E_k D_k}{\sum_{k \in g} E_k}$$

where `E_k` is SCB employment and `D_k` is the DAIOE metric for each 4-digit child `k`. Missing DAIOE values are excluded from both numerator and denominator.

Simple (unweighted) aggregation uses the arithmetic mean across the available children:

$$D^{(s)}_g = \frac{1}{|S_g|} \sum_{k \in S_g} D_k$$

where `S_g` contains every child with a non-null DAIOE score. Percentile ranks are computed per `(year, level)` slice after stacking all levels, giving comparable order statistics per metric.
